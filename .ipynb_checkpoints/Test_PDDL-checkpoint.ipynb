```python
from PDDL import PDDL_Parser
parser = PDDL_Parser()
domain = '/Users/Moi/Desktop/Centrale/3A/Planification/PowerPlanner/domains/domain_recipies.pddl'
problem = '/Users/Moi/Desktop/Centrale/3A/Planification/PowerPlanner/domains/problems/pb_tartiflette.pddl'
parser.parse_domain(domain)
parser.parse_problem(problem)
print('Domain name: ' + parser.domain_name)
for act in parser.actions:
    print(act.__repr__())
```

```python
from pddlpy import DomainProblem
```

```python
domainfile = "./domains/domain_recipies.pddl" 
problemfile = "./domains/problems/pb_tartiflette.pddl" 
domprob = DomainProblem(domainfile, problemfile)
```

```python
var_types
```

```python
import itertools
op = 'cook'
operator = domprob.domain.operators[op]
var_types = operator.variable_list.values()
var_combinations = itertools.product(*[reversed_objects[typ] for typ in var_types])
for var_tuple in var_combinations:
    print(operator.precondition_pos)
```

```python
var = domprob.domain.operators['put'].__dict__
print(var)
from collections import defaultdict
reversed_objects = defaultdict(list)
for obj,typ in domprob.problem.objects.items():
    reversed_objects[typ].append(obj)
```

```python
L = list(domprob.ground_operator('put'))
for i in range(len(L)):
    print(L[i].precondition_pos)
    print(L[i].effect_pos)
    print(L[i].effect_neg)
```

```python
ops_to_test = {1:"put", 2:"move", 3:"move", 4:"move", 5:"cut"}
op = ops_to_test[1]
print("ground for operator", op, "applicable if (adjacent loc1 loc2)")
for o in domprob.ground_operator(op):
    if ("adjacent","loc1","loc2") in o.precondition_pos:
        print()
        print( "\tvars", o.variable_list )
        print( "\tpre+", o.precondition_pos )
        print( "\tpre-", o.precondition_neg )
        print( "\teff+", o.effect_pos )
        print( "\teff-", o.effect_neg )
```
